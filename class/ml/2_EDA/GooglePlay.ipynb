```python
df["Size"].unique() # some are in kb and some are in mb and "varies with device "
```

```python
# 1 Mb= 1024 kb
def size_process(item):
    if(str(item)[-1]=="M"):
        res=float(str(item).replace("M",""))*1024
        return res
    elif(str(item)[-1]=="K"):
        res=float(str(item).replace("K",""))
        return res
    else:
        return str(np.NaN)
    
        
```

```python
df_copy.Size=df_copy["Size"].apply(size_process)
```
