```python
#merge
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
df1
```

```python
df2 =  pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
) 
df2
```

```python
# inner merge
pd.merge(df1,df2,how="inner")
```

```python
# left merge
pd.merge(df1,df2,how="left")
```

```python
# right
pd.merge(df1,df2,how="right")
```

```python
pd.merge(df1, df2, how = "outer") #all the common element of the common column from both the dataframe
```

```python
pd.merge(df1, df2, how = "cross")
```

```python
pd.merge(df1, df2, how = 'left', left_on="key2", right_on="key4")
```

```python
#join >> join works based on index/merge works based on columns
```

```python
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                    index = ['a', 'b', 'c', 'd', 'e']
)
df1
```

```python
df2 =  pd.DataFrame({'key6':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
},
index = ['a', 'b', 'h', 'i', 'j']
)
df2
```

```python
df1.join(df2, how = "inner") #joins columns of another dataframe based on common index
```

```python
df1.join(df2, how = "left")
```

```python
df1.join(df2, how = "right")
```

```python
df1.join(df2, how = "outer")
```

```python
df1.join(df2, how = "cross")
```
